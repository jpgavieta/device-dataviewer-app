# UI Data Filters: 
Functions that require user input to filter and visualize the loaded data. 

## Clean Data Toggle


```python
import sys
import os
sys.path.insert(0, os.path.abspath(".."))  # adds dataDashboard/ to path

import pandas as pd
from src.etl import get_data, clean_data

# Upload file (simulate file upload button)
FILE_UPLOAD = "sample_data/atmodat_old.csv"
with open(FILE_UPLOAD, "rb") as f:
    file_bytes = f.read()

data = get_data(file_bytes, device_type="Atmotube Pro/Pro 2")

# Simulate selecting pm1, pm2.5, and press_hpa
selected = {
    "pm":      ["pm1_0_ugm3_atm", "pm2_5_ugm3_atm", "pm10_ugm3_atm"],
    "weather": ["press_hpa"],
    "gas":     [],
    "phone":   [],
    "sat":     [],
    "txt":     [],
}

# Before clean
print("=== BEFORE ===")
for key in ["pm", "weather"]:
    df = data["data"][key]["df"]
    print(f"{key}: {len(df)} rows, NaNs: {df[selected[key]].isna().sum().to_dict()}")

# After clean
cleaned = clean_data(data["data"], selected)

print("\n=== AFTER ===")
for key in ["pm", "weather"]:
    df = cleaned[key]["df"]
    print(f"{key}: {len(df)} rows, NaNs: {df[selected[key]].isna().sum().to_dict()}")

# Confirm row counts match across dfs
print("\n=== ROW COUNT SYNC CHECK ===")
for key in cleaned:
    print(f"{key}: {len(cleaned[key]['df'])} rows")
```
